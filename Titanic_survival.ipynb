#### Pclass : 1 = 1st, 2 = 2nd, 3 = 3rd

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.iloc[:,1:-1]
X
```

```python
y = df.iloc[:,-1].values
```

```python
X = X.apply(LabelEncoder().fit_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Class 'Models' to define and save the models used

```python
class Models:
        
    def __init__(self,name,accuracy):
        self.name = name
        self.accuracy = accuracy
        
    def output(self):
        return f'Name of model is {self.name} and accuracy is {self.accuracy}'
```

## List to save all models

```python
mod_list = []
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0,max_iter=1000, solver='lbfgs',multi_class='ovr')
```

```python
clf.fit(X_train,y_train)
```

```python
lg_pred = clf.predict(X_test)
```

```python
 from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,lg_pred)
```

```python
print('Accuracy of this model is',acc*100)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,lg_pred)
```

```python
mod_list.append(Models('Logistic Regression',acc*100))
```

## SVC (Simple Vector Classification)

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto',kernel = 'rbf')
```

```python
clf.fit(X_train,y_train)
```

```python
svc_pred = clf.predict(X_test)
```

```python
acc = accuracy_score(y_test,svc_pred)
```

```python
print('Accuracy is ',acc*100)
```

```python
confusion_matrix(y_test,svc_pred)
```

```python
mod_list.append(Models('SVC',acc*100))
```

## Naive-Bayes Classification

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
```

```python
clf.fit(X_train,y_train)
```

```python
nb_pred = clf.predict(X_test)
```

```python
acc = accuracy_score(y_test,nb_pred)
```

```python
print('Accuracy is ',acc*100)
```

```python
confusion_matrix(y_test,nb_pred)
```

```python
mod_list.append(Models('Naive-Bayes',acc*100))
```

## Analysing performance of all models used

```python
for models in mod_list:
    print(models.output())
```
